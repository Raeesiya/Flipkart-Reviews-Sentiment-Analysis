### WordCloud

```python
from wordcloud import WordCloud
# Combine all reviews into a single text for the word cloud
text_data = " ".join(df['review'].astype(str).tolist())

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(text_data)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Word Cloud of Reviews", fontsize=16)
plt.show()
```
